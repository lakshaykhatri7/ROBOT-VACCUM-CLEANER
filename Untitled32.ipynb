```python
import numpy as np
import matplotlib.pyplot as plt
import random
import json
import torch
import torch.nn as nn
import torch.optim as optim
from collections import deque
```

```python
def generate_dataset(num_samples=50, grid_size=5):
    dataset = []
    
    for _ in range(num_samples):
        grid = np.zeros((grid_size, grid_size))
        for _ in range(random.randint(3, 10)):
            x = random.randint(0, grid_size - 1)
            y = random.randint(0, grid_size - 1)
            grid[x][y] = 1
        dataset.append(grid.tolist())
    
    return np.array(dataset)

dataset = generate_dataset()
```

```python
class VacuumEnv:
    def __init__(self, grid):
        self.grid = np.array(grid)
        self.size = self.grid.shape[0]
        self.reset()

    def reset(self):
        self.agent_pos = [0, 0]
        self.active_grid = self.grid.copy()
        return tuple(self.agent_pos)

    def step(self, action):
        x, y = self.agent_pos
        
        if action == 0 and x > 0: x -= 1
        if action == 1 and x < self.size-1: x += 1
        if action == 2 and y > 0: y -= 1
        if action == 3 and y < self.size-1: y += 1
        
        self.agent_pos = [x, y]
        
        reward = 0
        if self.active_grid[x][y] == 1:
            reward = 10
            self.active_grid[x][y] = 0
        
        done = np.sum(self.active_grid) == 0
        return (x, y), reward, done
```

```python
class QLearningAgent:
    def __init__(self, grid_size):
        self.q_table = np.zeros((grid_size, grid_size, 4))
        self.lr = 0.1
        self.gamma = 0.9
        self.epsilon = 0.3

    def choose_action(self, state):
        if random.random() < self.epsilon:
            return random.randint(0, 3)
        return np.argmax(self.q_table[state])

    def update(self, state, action, reward, next_state):
        max_future = np.max(self.q_table[next_state])
        cur_q = self.q_table[state][action]
        self.q_table[state][action] = cur_q + self.lr * (reward + self.gamma * max_future - cur_q)
```

```python
q_rewards = []

for episode in range(len(dataset)):
    env = VacuumEnv(dataset[episode])
    state = env.reset()
    done = False
    total_reward = 0
    
    q_agent = QLearningAgent(5)
    
    while not done:
        action = q_agent.choose_action(state)
        next_state, reward, done = env.step(action)
        q_agent.update(state, action, reward, next_state)
        state = next_state
        total_reward += reward
        
    q_rewards.append(total_reward)
```

```python
class DQN(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc = nn.Sequential(
            nn.Linear(2, 64),
            nn.ReLU(),
            nn.Linear(64, 4)
        )

    def forward(self, x):
        return self.fc(x)
```

```python
dqn = DQN()
optimizer = optim.Adam(dqn.parameters(), lr=0.001)
loss_fn = nn.MSELoss()
memory = deque(maxlen=10000)

def get_action(state, epsilon=0.2):
    if random.random() < epsilon:
        return random.randint(0, 3)
    state = torch.FloatTensor(state)
    qvals = dqn(state)
    return torch.argmax(qvals).item()

dqn_rewards = []

for episode in range(len(dataset)):
    env = VacuumEnv(dataset[episode])
    state = env.reset()
    done = False
    total_reward = 0

    while not done:
        action = get_action(state)
        next_state, reward, done = env.step(action)
        
        memory.append((state, action, reward, next_state, done))
        
        state = next_state
        total_reward += reward
        
        if len(memory) > 32:
            batch = random.sample(memory, 32)
            states, actions, rewards, next_states, dones = zip(*batch)

            states = torch.FloatTensor(states)
            actions = torch.LongTensor(actions)
            rewards = torch.FloatTensor(rewards)
            next_states = torch.FloatTensor(next_states)
            dones = torch.FloatTensor(dones)

            q_values = dqn(states).gather(1, actions.unsqueeze(1)).squeeze()
            next_q_values = dqn(next_states).max(1)[0]
            targets = rewards + 0.9 * next_q_values * (1 - dones)

            loss = loss_fn(q_values, targets.detach())
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

    dqn_rewards.append(total_reward)
```

```python
def show_grid(grid, pos):
    grid_copy = grid.copy()
    grid_copy[pos[0]][pos[1]] = 9
    plt.imshow(grid_copy, cmap="cool")
    plt.title("Simulation (9 = Agent)")
    plt.show()

sample = dataset[0]
env = VacuumEnv(sample)
show_grid(sample, env.agent_pos)
```

```python
plt.plot(q_rewards, label="Q-Learning")
plt.plot(dqn_rewards, label="DQN")
plt.legend()
plt.xlabel("Episode")
plt.ylabel("Total Reward")
plt.title("Reward Comparison")
plt.show()
```

```python
def smooth(x, w=10):
    return np.convolve(x, np.ones(w)/w, mode='valid')

plt.plot(smooth(q_rewards), label="Q-Learning Smoothed")
plt.plot(smooth(dqn_rewards), label="DQN Smoothed")
plt.legend()
plt.show()
```
